```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Cleaning

```python
df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis = 1,inplace =True)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace =  True)
```

```python
df.shape
```

```python
df['rate'].unique()
```

```python
df.rate.isnull().sum()
```

```python
def rate1(value):
    if value == 'NEW' or value == '-':
        return  np.nan
    else:
        value = str(value).split(' / ')
        value = value[0]
        return float(value)
```

```python
df['rate'] = df['rate'].apply(rate1)
```

```python
df['rate'].unique()
```

```python
df.head()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
df.rate.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
 df.info()
```

```python
 df.rename(columns= {'approx_cost(for two people)' : 'cost2plates','listed_in(type)'  : 'Type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df = df.drop(['listed_in(city)'], axis =1)
```

```python
df['cost2plates'].unique()
```

```python
def handelcomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['cost2plates'] = df['cost2plates'].apply(handelcomma )
df['cost2plates'].unique()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handel_rest(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else: 
        return value
df['rest_type'] = df['rest_type'].apply(handel_rest)
df['rest_type'].value_counts()
    
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location <300]
def handle_location(value):
    if (value in location_lessthan300):
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines_lessthan100 = cuisines[cuisines<100]
def handle_cuisines(value):
    if (value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
